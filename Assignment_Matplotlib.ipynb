Problem Statement 1 : We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1. fitting it to the periodic function
2. plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pandas as pd
```

```python
t_max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
t_min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
mnth = np.arange(12)
plt.plot(mnth, t_max, 'ro')
plt.plot(mnth, t_min, 'bo')
plt.xlabel('Month')
plt.ylabel('Temperature(Â°C)')
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, mnth,
                                      t_max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, mnth,
                                      t_min, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(mnth, t_max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(mnth, t_min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature(Â°C)')
plt.show()
```

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
df=pd.read_csv(url)
```

```python
# 1. Create a pie chart presenting the male/female proportion
plt.pie(df['sex'].value_counts(),labels = ['Male','Female'],autopct='%1.1f%%')
plt.title('Male female proportion')
```

```python
#2.Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
df = df.dropna(subset=['sex'])
color={'female':'blue','male':'orange'}
plt.scatter(df['fare'],df['age'],c=df.sex.map(color))
plt.xlabel('fare') 
plt.ylabel('age')
plt.title('Fair paid and Age')
```

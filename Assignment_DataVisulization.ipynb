Problem Statement  : In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
import numpy as  np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn import decomposition
import pandas as pd
```

```python
#load iris dataset from sklearn
iris=datasets.load_iris()
iris
```

```python
#Iris data
X=iris.data
X
```

```python
#Iris targets
y=iris.target
y
```

```python
iris.data.shape
```

```python
#dimensionality reduction of iris data using pca(4 to 3)
d=decomposition.PCA(3)
id=d.fit_transform(X)
id

```

```python
df=pd.DataFrame(data=id,columns=['X','Y','Z']) #convert id to df
X,Y,Z=np.array(df['X']) ,np.array(df['Y']),np.array(df['Z'])
```

```python

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.scatter3D(X,Y,Z,c=iris.target,cmap='brg')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
        
plt.title('Iris dataset with Classes Setosa as Blue,Versicolor as Red,  Virginica as Green')
plt.show()
```
